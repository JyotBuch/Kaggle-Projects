```python
#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing the dataset
dataset = pd.read_csv("master.csv")
dataset.head()
```

```python
suicideyears = dataset.groupby(['year'])['suicides_no'].sum()
suicideyears.plot(figsize=(8,6), linewidth=2, fontsize=15,color='red')
plt.xlabel('Year', fontsize=15)
plt.ylabel('Number of Suicides',fontsize=15)
plt.title("Number of Suicides over the years 1985-2015", fontsize=20)
```

From the above graph, we can conclude that the number of suicides first increased and then significantly decreased over the years

```python
plt.figure(figsize=(13,6))
sns.barplot(x=dataset['age'], y='suicides_no', hue= 'sex', data= dataset)
plt.xlabel('Age', fontsize= 15)
plt.ylabel('Number of suicides', fontsize= 15)
plt.title("Number of suicides between different age groups", fontsize = 20)
```

```python
plt.figure(figsize=(13,6))
sns.barplot(x=dataset['suicides_no'], y='sex', data= dataset)
plt.xlabel('Number of suicides', fontsize= 15)
plt.ylabel('Sex', fontsize= 15)
plt.title("Number of suicides between different Genders", fontsize = 20)
```

Over here, we can see how the number of males commiting suicides is more than the number of females

```python
plt.figure(figsize=(50,10))
sns.lineplot(x='year', y='suicides_no', hue='age', data= dataset)
plt.xlabel("Year" , fontsize = 25)
plt.ylabel("Number of Suicides", fontsize = 25)
plt.title("Number of suicides vs Years separated based on different Ages", fontsize = 50)
```

We can also observe that over the years, the older age group(35-54 years), have a higher chance of commiting suicide

```python
plt.figure(figsize=(8,6))
sns.barplot(x='sex', y='suicides_no', hue='generation', data= dataset)
plt.xlabel("Genders" , fontsize = 15)
plt.ylabel("Number of Suicides", fontsize = 15)
plt.title("Number of suicides vs Gender separated based on different Generations", fontsize = 20)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(dataset.corr(), annot=True)
plt.title("Heatmap", fontsize=20)
```

From the above heatmap, we can see that there is no linear relationship between GDP per capita/HDI per year and Suicide rate so we can say for this data, GDP or HDI do not play a major role in increasing number of suicides.
